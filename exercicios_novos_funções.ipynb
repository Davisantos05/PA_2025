Gerar senha


```python
import random
import string

def gerar_senha():
    maiuscula = random.choice(string.ascii_uppercase)
    minuscula = random.choice(string.ascii_lowercase)
    numero = random.choice(string.digits)
    especial = random.choice(string.punctuation)
    outros = ''.join(random.choice(string.ascii_letters + string.digits + string.punctuation) for _ in range(4))
    senha = maiuscula + minuscula + numero + especial + outros
    senha = ''.join(random.sample(senha, len(senha)))
    return senha
```

validar senha

```python
import string

def validar_senha(senha):
    resultado = {
        "tamanho_valido": len(senha) >= 8,
        "tem_numero": any(c.isdigit() for c in senha),
        "tem_maiuscula": any(c.isupper() for c in senha),
        "tem_minuscula": any(c.islower() for c in senha),
        "tem_especial": any(c in string.punctuation for c in senha),
    }
    resultado["senha_valida"] = all(resultado.values())
    return resultado
```
